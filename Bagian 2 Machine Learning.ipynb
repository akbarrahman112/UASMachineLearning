```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
dataset = pd.read_csv('../ML/Liga120192021.csv')
```

```python
dataset.tail()
```

```python
dataset.keys()
```

```python
dataku = pd.DataFrame(dataset)
dataku.head()
```

```python
X = np.asarray(dataset)
print (X)
```

```python
plt.scatter(X[:,0],X[:,1], label='True Position')
plt.xlabel("Pass9")
plt.ylabel("Pass10")
plt.title("Grafik Passing 9-10")
plt.show()
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
print(kmeans.cluster_centers_)
```

```python
print(kmeans.labels_)
```

```python
plt.scatter(X[:,0],X[:,1], c=kmeans.labels_, cmap='rainbow')
plt.xlabel("Pass9")
plt.ylabel("Pass10")
plt.title("Grafik Passing 9-10")
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='black')
plt.xlabel("Pass9")
plt.ylabel("Pass10")
plt.title("Grafik Passing 9-10")
plt.show()
```
